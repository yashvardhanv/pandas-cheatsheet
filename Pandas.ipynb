###**important methods for updating rows**
1.apply() 
2.map() 
3.applymap()

```python
# Practicing Problems: total survivers, male survivers and % of males who survived in New dataframe
Male_sur= df.loc[df["Sex"]=="male","Survived"].value_counts()
# total_sur=pd.DataFrame(df['Survived'].value_counts())................... similar as the line below
total_sur = pd.DataFrame(GrpWhoSur['Sex'].apply(lambda x: x.str.contains('male').sum()))
total_sur.columns = total_sur.columns.str.replace('Sex', 'Total_sur')
New_Df= pd.concat([total_sur,Male_sur],axis='columns',sort=False)
New_Df.rename(columns={'Survived':  'Male_sur'}, inplace=True)
New_Df['Percentage'] = (New_Df['Male_sur']/New_Df['Total_sur']) * 100
New_Df.loc[1]
```

##Cleaning Data:

```python
#for handling NaN values
import numpy as np
```

```python
people = {
    'first': ['Yash', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Vyas', 'Doe', 'Doe', 'Bumstead', np.nan, np.nan, 'Missing'], 
    'email': ['Yashvardhan@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['19', '55', '63', '26', None, None, 'Missing']
}
dFromDict = pd.DataFrame(people)
```

```python
dFromDict
```

```python
dFromDict = pd.DataFrame(people)
#replace customized nan values with np.NaN
dFromDict.replace('NA', np.nan, inplace=True)
dFromDict.replace('Missing', np.nan, inplace=True)
```

```python
# drops column or row if it contains NaN
#index = rows, Columns = drop column
dFromDict.dropna(axis='index', how='any', subset=['last', 'email'])
```

```python
# returns true if element is NaN
dFromDict.isna()
```

```python
# fills NaN values with parameter specified
# can use 0 for numeric data IE: if a student is absent on test then marks will be 0 instead of NaN
dFromDict.fillna("Missing")
```

```python
# changing type of columns
# age was string before now its Float
# only nan values are allowed no Strings value is allowed otw error is thrown 
dFromDict['age'] = dFromDict['age'].astype(float)
dFromDict['age'].mean()
```
